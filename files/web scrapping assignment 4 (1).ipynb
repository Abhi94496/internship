```python
import numpy as np
import pandas as pd
from selenium import  webdriver
import selenium.webdriver.common.by as by
import warnings
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
'''1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views'''

# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

# Open the Wikipedia page with the list of most-viewed YouTube videos
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

# Maximize the window for better visibility of elements
driver.maximize_window()

# Pause for 5 seconds to allow the page to load
time.sleep(5)

# Extract details using Xpath selectors
#name
content = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
name = [i.text for i in content]

#Artist
content1 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
artist = [i.text for i in content1]

#views
content2 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
views = [i.text for i in content2]

#upload date
content3 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
date = [i.text for i in content3]

# Generating rank numbers from 1 to the length of the 'name' list
rank = [i for i in range(1, len(name) + 1)]

# Create a DataFrame to store the scraped data
data = pd.DataFrame({
    'Rank': rank,
    'Name': name,
    'Artist': artist,
    'Upload Date': date,
    'Views (Billions)': views
})

data
```

```python
''' 2 )Scrape the details team India’s international fixtures from bcci.tv. 
Url = https://www.bcci.tv/. 
You need to find following details: 
A) Series 
B) Place 
C) Date 
D) Time'''

driver2 = webdriver.Chrome()
# Open the BCCI website
driver2.get('https://www.bcci.tv/')

# Maximize the window for better visibility of elements
driver2.maximize_window()
time.sleep(5)

#click on FIXTURE option
click1 =  driver2.find_element(By.XPATH, '//div[@class="imw-tabs international-tabs"]/a[2]')
click1.click()
time.sleep(8)

# Scroll to the element with the upcoming matches
click = driver2.find_element(By.XPATH, '//div[@class="col-lg-12 col-md-12 col-sm-12 ng-scope"][9]')
driver2.execute_script("arguments[0].scrollIntoView();", click)
time.sleep(10)

# Click on the "More Matches" button
click2 = driver2.find_element(By.XPATH, '//button[@class="match-btn btn-red d-flex align-items-center justify-content-center mx-auto mt-3 morematches"]')
click2.click()
time.sleep(5) 

# Extract details using Xpath selectors
#series
series = driver2.find_elements(By.XPATH,'//div[@class="match-card-top"]/div/h5 |//div[@class="match-card-middle d-flex"]/h5')
ser = [i.text for i in series]

#place
pla = driver2.find_elements(By.XPATH,'//div[@class="match-place ng-scope"] | //div[@class="match-venue ng-scope"]')
place = [i.text for i in pla]

#date
da = driver2.find_elements(By.XPATH, '//div[@class="match-date-info"]/div[1]')
date = [i.text for i in da]

#time
ti = driver2.find_elements(By.XPATH, '//div[@class="match-date-info"]/div[2]')
times = [i.text for i in ti]

# Create a Pandas DataFrame
data2 = pd.DataFrame({'series' : ser, 'place': place, 'date': date, 'time' : times })
data2
```

```python
''' 3) Scrape the details of State-wise GDP of India from statisticstime.com. 
Url = http://statisticstimes.com/ 
You have to find following details: A) Rank 
B) State 
C) GSDP(18-19)- at current prices 
D) GSDP(19-20)- at current prices 
E) Share(18-19) 
F) GDP($ billion)'''

'''NOTE : in the given website 'GSDP(18-19)- at current prices' , 'GSDP(19-20)- at current prices' and 'Share(18-19)' 
are not available  and they are updated to 'GSDP(21-22)- at current prices' , 'GSDP(22-23)- at current prices' and 'Share(21-22)' '''

driver3 = webdriver.Chrome()
#opening twice to avoid the google ads
for a in range(2):
    driver3.get('http://statisticstimes.com/')
    time.sleep(5)
#click on economy button
    click3 = driver3.find_element(By.XPATH,'//div[@class="navbar"]/div[2]/button')
    click3.click()
#click on INDIA button
    click2 = driver3.find_element(By.XPATH,'//div[@class="navbar"]/div[2]/div/a[3]')
    click2.click()
#click on GDP ON INDIAN STATES option
click4 = driver3.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
click4.click()

# Finding elements using Xpath for different columns in the table\
#rank
rank = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[1]')
ranks = [i.text for i in rank]

#state
state = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[2]')
states = [i.text for i in state]

#  GSDP(18-19)- at current prices
gsdp19 = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[3]')
gsdps19 = [i.text for i in gsdp19]

# GSDP(19-20)- at current prices 
gsdp20 = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[4]')
gsdps20 = [i.text for i in gsdp20]

#Share(18-19) 
share = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[5]')
shares = [i.text for i in share]

#GDP($ billion)
gdp = driver3.find_elements(By.XPATH, '//div[@class="fwidth"][3]/div/div/table/tbody/tr/td[6]')
gdps = [i.text for i in gdp]

# Create a Pandas DataFrame
data3 = pd.DataFrame({'Rank' : ranks, 'State': states, ' GSDP(21-22)- at current prices ' : gsdps19, 'GSDP(22-23)- at current prices'
                      : gsdps20, 'share (21-22)' : shares, 'GDP($ billion)': gdps  })
data3
```

```python
'''4) Scrape the details of trending repositories on Github.com. 
Url = https://github.com/ 
You have to find the following details: 
A) Repository title 
B) Repository description 
C) Contributors count 
D) Language used'''

driver4 = webdriver.Chrome()
driver4.get('https://github.com/')
time.sleep(5)
#click on OPEN SOURCE option
click = driver4.find_element(By.XPATH, '//ul[@class="d-lg-flex list-style-none"]/li[3]')
click.click()

#click on TRENDING option
click1 = driver4.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
click1.click()
time.sleep(5)


# Lists to store extracted information
r_titles = []
r_dis = []
r_lan = []

# Find all <article> elements on the page
articles = driver4.find_elements(By.XPATH, '//article[@class="Box-row"]')

# Loop through each article
for i in range(1, len(articles)+1):  
    # Get the current article based on index i
    article = driver4.find_element(By.XPATH, f'(//article[@class="Box-row"])[{i}]')
    
    # Find the element for <h2> (title) within the current article
    title_ = article.find_element(By.XPATH, './/h2')
    r_titles.append(title_.text)
    
    try:
        # Find the element for <p> (description) within the current article
        description = article.find_element(By.XPATH, './/p')
        description_text = description.text if description else '---'
    except NoSuchElementException:
        description_text = '---'
    r_dis.append(description_text)

    try:
        # Find the element for programming language <span>
        language = article.find_element(By.XPATH, './/span[@itemprop="programmingLanguage"]')
        language_text = language.text.strip() if language else '---'
    except NoSuchElementException:
        language_text = '---'
    r_lan.append(language_text)

# Find contributors count using XPATH
cc = driver4.find_elements(By.XPATH, '//div[@class="f6 color-fg-muted mt-2"]/a[2]')
contributors = [i.text for i in cc]

# Create a Pandas DataFrame
data4 = pd.DataFrame({
    'Repository title': r_titles,
    'Repository description': r_dis,
    'Contributors count': contributors,
    'Language used': r_lan})

data4
```

```python
'''5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the 
following details: 
A) Song name 
B) Artist name 
C) Last week rank 
D) Peak rank 
E) Weeks on board'''

# Create a new instance of the Chrome driver
driver5 = webdriver.Chrome()

# Open the Billboard website
driver5.get('http://www.billboard.com/')

# Click on the "Charts" option in the navigation menu
click =  driver5.find_element(By.XPATH,'//div[@class="lrv-u-flex lrv-u-flex-grow-1 lrv-u-flex-basis-60p"]/div/div/nav/ul/li[1]')
click.click()

# Click on the "Hot 100" option in the submenu
click1 = driver5.find_element(By.XPATH, '//div[@class="u-flex-grow-0 lrv-u-text-align-center"]/span/a')
click1.click()

# Extract details using Xpath selectors
#song name
name =  driver5.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[1]/h3')
names = [i.text for i in name]

#Artist name
sing =  driver5.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[1]/h3')
singer = [i.text for i in sing]

#last week rank
lw =  driver5.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[4]/span')
lwr = [i.text for i in lw]

#peak rank
pr = driver5.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[5]/span')
pwr = [i.text for i in pr]

#weeks on board
wb = driver5.find_elements(By.XPATH, '//li[@class="lrv-u-width-100p"]/ul/li[6]/span')
wbr =  [i.text for i in wb]

# Create a DataFrame to store the scraped data
data5 = pd.DataFrame({
    'Song name' : names,
    'Artist name' : singer,
    'Last week rank' : lwr,
    'Peak rank' : pwr,
    'Weeks on board': wbr
})

data5
```

```python
''' 6) Scrape the details of Highest selling novels. 
A) Book name 
B) Author name 
C) Volumes sold 
D) Publisher 
E) Genre 
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'''
# Create a new instance of the Chrome driver
driver6 = webdriver.Chrome()

# Open The Guardian website to scrape information about best-selling books
driver6.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')

# Pause for 5 seconds to allow the page to load
time.sleep(5)

# Extract details using Xpath selectors

#book name
elem = driver6.find_elements(By.XPATH, '//div[@class="embed block"]/table/tbody/tr/td[1]')
book = [i.text for i in elem]

#author_name
author = driver6.find_elements(By.XPATH, '//div[@class="embed block"]/table/tbody/tr/td[2]')
author_name = [i.text for i in author]

#volumes sold
vol = driver6.find_elements(By.XPATH, '//div[@class="embed block"]/table/tbody/tr/td[3]')
volume_sold = [i.text for i in vol]

#publisher
publi = driver6.find_elements(By.XPATH, '//div[@class="embed block"]/table/tbody/tr/td[4]')
publisher = [i.text for i in publi]

#genre
geners = driver6.find_elements(By.XPATH, '//div[@class="embed block"]/table/tbody/tr/td[5]')
genre = [i.text for i in geners]

# Create a DataFrame to store the scraped data
data6 = pd.DataFrame({
    'Book name': book,
    'Author name': author_name,
    'Volume sold': volume_sold,
    'Publisher': publisher,
    'Genre': genre
})

data6
```

```python
''' 7) Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes'''


'''note : the given url //www.imdb.com/list/ls095964455/ is outdated.
so i scrapped the same details from updated https://www.imdb.com/list/ls512407256/'''

driver7 = webdriver.Chrome()

# Open the IMDb website to scrape information about a movie list
driver7.get('https://www.imdb.com/list/ls512407256/')

# Pause for 5 seconds to allow the page to load
time.sleep(5)

# Extract details using Xpath selectors
#year span
year = driver7.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/span[2]')
span = [i.text for i in year]

#name
name = driver7.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')
names = [i.text for i in name]

#run time
run = driver7.find_elements(By.XPATH, '//p[@class="text-muted text-small"]/span[3]')
r_time = [i.text for i in run]

#gener
gener = driver7.find_elements(By.XPATH, '//p[@class="text-muted text-small"]/span[5]')
geners = [i.text for i in gener]

#ratings
rate = driver7.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]/span[2]')
rates = [i.text for i in rate]

#votes
vote = driver7.find_elements(By.XPATH, '//p[@class="text-muted text-small"][3]/span[2]')
votes = [i.text for i in vote]

# Create a DataFrame to store the scraped data
data7 = pd.DataFrame({
    'Name': names,
    'Year span': span,
    
    'Genre': geners,
    'Run time': r_time,
    'Ratings': rates,
    'Votes': votes
})

data7
```

```python
'''8)Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ You 
have to find the following details: 
A) Dataset name 
B) Data type 
C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute 
G) Year'''

#Create a new instance of the Chrome driver
driver8 = webdriver.Chrome()

# Open the UCI Machine Learning Repository website
driver8.get('https://archive.ics.uci.edu/')

# Pause for 5 seconds to allow the page to load
time.sleep(5)

# Click on the first dataset link
click = driver8.find_element(By.XPATH, '//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
click.click()

# Pause for 5 seconds to allow the page to load
time.sleep(5)

# Click on the "Data Folder" label to view the data files
click1 = driver8.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
click1.click()

# Extract details using Xpath selectors
#dataset names
name = driver8.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div/div[2]/h2/a')
names = [i.text for i in name]

#data type
data_t = driver8.find_elements(By.XPATH, '//div[@class="rounded-box bg-base-100"]/div/div[2]/p')
type = [i.text for i in data_t]


#task, attribute type, number of instance, number of attribute
eleme = driver8.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div/span')
task = [i.text for i in eleme[0::4]]
attribute_type = [i.text for i  in eleme[1::4]]
no_instance = [i.text for i in eleme[2::4]]
no_attribute = [i.text for i in eleme[3::4]]

#year
ye = driver8.find_elements(By.XPATH, '//tbody[@class="border"]/tr/td[3]')
year =  [i.text for i in ye]

# Create a DataFrame to store the scraped data
data8 = pd.DataFrame({
    'Dataset_names': names,
    'Data type': type,
    'Task': task,
    'Attribute type': attribute_type,
    'No. of Instances': no_instance,
    'No. of Attributes': no_attribute,
    'Year': year
})
data8
```
