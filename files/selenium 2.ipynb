```python
'''1) Write a python program which searches all the product under a particular product from www.amazon.in. The 
product to be searched will be taken as input from user. For e.g. If user input is guitar. Then search for 
guitars. '''
driver = webdriver.Chrome()
time.sleep(5)

driver.get('http://www.amazon.in/')
time.sleep(5)

product = input("Enter the product you want: ")
time.sleep(10)

search = driver.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search.send_keys(product)
time.sleep(5)

click = driver.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
click.click()
time.sleep(5)

product_links = []

for page in range(1, 4):
    links = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in links:
        product_links.append(i.get_attribute('href'))

    if page == 1:
        click1 = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        click1.click()
    else:
        click1 = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        click1.click()

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')))

    time.sleep(5)

product_links
```

```python
''' 6) Write a program to scrap all the available details of best gaming laptops from digit.in. '''
# Set up the Chrome webdriver
driver = webdriver.Chrome()
driver.maximize_window()

# Open the digit.in website
driver.get('https://www.digit.in/')
time.sleep(3)

# Lists to store laptop details
laptop = []
urls = []
operating_system = []
display_size = []
resolution = []
processor = []

# Find URLs for the best gaming laptops
url_elements = driver.find_elements(By.XPATH, '//li[@class="rh-subitem-menus rh-megamenu menu-item menu-item-type-custom menu-item-object-custom menu-item-has-children"][2]/ul/li[3]/ul/li/a')
for i in url_elements:
    urls.append(i.get_attribute('href'))

# Loop through each URL to extract laptop details
for url in urls:
    driver.get(url)
    
    # Extract laptop names
    product_elements = driver.find_elements(By.XPATH, '//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]/a')
    for element in product_elements:
        laptop.append(element.text)
    
    # Extract details such as operating system, display size, resolution, and processor
    name_elements = driver.find_elements(By.XPATH, '//span[@class="meta_v_value"]')
    for i in name_elements[0::4]:
        operating_system.append(i.text)
    for i in name_elements[1::4]:
        display_size.append(i.text)
    for i in name_elements[2::4]:
        resolution.append(i.text)
    for i in name_elements[3::4]:
        processor.append(i.text)

# Create a DataFrame from the collected data
data6 = pd.DataFrame({
    'laptop': laptop,
    'operating_system': operating_system,
    'display_size': display_size,
    'resolution': resolution,
    'processor': processor
})

data6
```

```python
'''8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
from any YouTube Video. 
'''

# Set up the Chrome webdriver and open the YouTube video URL
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/watch?v=O4ymI8QwEeg')
time.sleep(3)
driver.maximize_window()

# Scroll down to load comments
time.sleep(3)
driver.execute_script("window.scrollBy(0, 500);")
time.sleep(5)

# Lists to store comments, upvotes, and posted times
comments = []
upvotes = []
times = []

# Loop to extract at least 500 comments
while len(comments) < 500:
    # Extract comments
    comment_elements = driver.find_elements(By.XPATH, '//yt-formatted-string[@id="content-text"]')
    for element in comment_elements:
        comments.append(element.text)
    
    # Extract upvotes
    upvote_elements = driver.find_elements(By.XPATH, '//div[@class="style-scope ytd-comment-action-buttons-renderer"]/span')
    for element in upvote_elements[1::2]:
        upvotes.append(element.text)

    # Extract posted times
    time_elements = driver.find_elements(By.XPATH, '//div[@class="style-scope ytd-comment-renderer"]/yt-formatted-string/a')
    for element in time_elements:
        times.append(element.text)
    
    # Scroll further to load more comments
    driver.execute_script("window.scrollBy(0, 4000);")
    time.sleep(3)

# Create a DataFrame from the collected data
data8 = pd.DataFrame({
    'comments': comments,
    'upvotes': upvotes,
    'posted time': times})

data8
```
