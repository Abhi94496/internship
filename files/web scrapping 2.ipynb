```python
'''Q7: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then
set CPU Type filter to “Intel Core i7” as shown in the below image:
After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price'''

driver7 = webdriver.Chrome()
time.sleep(5)
driver7.get('https://www.amazon.in/')
time.sleep(10)

search = driver7.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search.send_keys('laptop')
time.sleep(3)

click = driver7.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
click.click()
time.sleep(10)

ratings = []
price1 = []
product7 = []

laptop = driver7.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal" or @class="a-size-base-plus a-color-base a-text-normal"]')
limit = 0
for i in laptop:
    if limit < 12:
        product7.append(i.text.strip())
        limit += 1
filtered_list = [item for item in product7 if item != ' ']
# Remove empty string (' ')
filtered_list = [item for item in product7 if item != ' ']
filtered_laptops = [laptop for laptop in filtered_list if laptop]


price_elements = driver7.find_elements(By.XPATH, '//span[@class="a-price-whole"]')
limit01 = 0
for element in price_elements:
    # Check if the text is not an empty string
    if limit01 < 10 and element.text.strip(): 
        # Additional condition to skip empty strings
        price1.append(element.text.strip())
        limit01 += 1
filtered_prices = [price for price in price1 if price]

# Get the HTML content of the page
page_html = driver7.page_source
# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(page_html, 'html.parser')
# Find all span elements with the specified class
rating_elements = soup.find_all('span', class_='a-icon-alt')
rating_elements
limit_r = 0
# Extract the text content of each span element
for element in rating_elements[2:]:
    if limit_r < 10:
        ratings.append(element.get_text(strip=True) )
        limit_r+=1

#creating data frame 
data7 = pd.DataFrame({'PRODUCT' : filtered_laptops , 'PRICE': filtered_prices , 'RATING' : ratings})
data7

#if error occurs run this again
```

```python
'''Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
'''
driver8 = webdriver.Chrome()
time.sleep(5)

#opening the azquote page on automated chrome browser
driver8.get('https://www.azquotes.com/')
time.sleep(5)

#clicking top quotes button
click = driver8.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
click.click()
time.sleep(5)

quotes= []
type = []
authors = []
qlimit = 0
alimit = 0
tlimit = 0
for k in range (0,10):
    
    # Code for scraping QUOTES
    quotes1 = driver8.find_elements(By.XPATH, '//a[@class="title"]')
    for i  in quotes1:
        if qlimit< 1000:
            quotes.append(i.text)
    
    # Code for scraping AUTHORS
    authors1 =driver8.find_elements(By.XPATH, '//div[@class="author"]')
    for i in authors1:
        if alimit < 1000:
            authors.append(i.text)
    
    # Code for scraping QUOTE TYPE
    type1 =driver8.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type1:
        if tlimit < 1000:
            type.append(i.text)
    print('page', k, 'scrapped')
    
    # Clicking on the next page  
    if k == 0:
        click81 = driver8.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
        click81.click()
        time.sleep(10)
    elif k > 0 and k < 9:
        click81 = driver8.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]/a')
        click81.click()
        time.sleep(10)
    else:
        print('scrapping is ended')

#creating data frame 
data8 = pd.DataFrame({'quotes' : quotes, 'type'  :type, 'authors' :authors})
data8
```
